```python
df['ASSDLANDVALUEPERTOTALLANDSQUAREFOOTAGE'] = pd.to_numeric(df['ASSDLANDVALUEPERTOTALLANDSQUAREFOOTAGE'], errors='coerce') 
```

```python
print(df.columns.tolist())
```

```python
# Creating legend categories here based on the categories above and input provided by Lenny
conditions = [
    (df['ASSDLANDVALUEPERTOTALLANDSQUAREFOOTAGE'] < 15),
    (df['ASSDLANDVALUEPERTOTALLANDSQUAREFOOTAGE'] > 15) & (df['ASSDLANDVALUEPERTOTALLANDSQUAREFOOTAGE'] <= 35),
    (df['ASSDLANDVALUEPERTOTALLANDSQUAREFOOTAGE'] > 35) & (df['ASSDLANDVALUEPERTOTALLANDSQUAREFOOTAGE'] <= 60),
    (df['ASSDLANDVALUEPERTOTALLANDSQUAREFOOTAGE'] > 60) & (df['ASSDLANDVALUEPERTOTALLANDSQUAREFOOTAGE'] <= 100),
    (df['ASSDLANDVALUEPERTOTALLANDSQUAREFOOTAGE'] > 100) & (df['ASSDLANDVALUEPERTOTALLANDSQUAREFOOTAGE'] <= 250),
    (df['ASSDLANDVALUEPERTOTALLANDSQUAREFOOTAGE'] > 250)
]
choices = ['1', '2', '3', '4', '5', '6']
default_value = '0'

# Create the new column using np.select()
df['LEGENDCATEGORY'] = np.select(conditions, choices, default=default_value)

df.head()
```

```python
df = df.dropna(subset=['ASSDLANDVALUEPERTOTALLANDSQUAREFOOTAGE'])
df.head()
```

```python
df = df.dropna(how='all')  # drops rows where every column is empty
```
